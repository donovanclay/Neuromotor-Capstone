```python
import sys
from io import StringIO
import os
import matplotlib.pyplot as plt
# import mathplot
import pandas as pd
import numpy as np

```

```python
file_path = os.path.abspath('RepositoryData/SL01-T01/joints.txt')
f = open(file_path, 'r')

lines = f.readlines()
lines = lines[2:]

col_names = ["time", "GHR", "GKR", "GAR", "GHL", "GKL", "GAL", "PHR", "PKR", "PAR", "PHL", "PKL", "PAL", "test"]

DATA = StringIO(''.join(lines))

df = pd.read_csv(DATA, names=col_names, sep="\t", header=None)
del df['test']


print(df[0:10])
```

```python
print(col_names[1:][:-1])
```

```python
time_array = df['time']

l1 = col_names[1:][:-1][0:6]
l2 = col_names[1:][:-1][6:12]

# my_range = range(22_000, 23_000)
# my_range = range(18_000, 50_000)
my_range = range(18_000, 25_000)


# graph_names = [val for pair in zip(l1, l2) for val in pair]
pairs = zip(l1, l2)
for pair in pairs:
    col1 = pair[0]
    col2 = pair[1]

    

    time = time_array[my_range].values
    true_values = df[col1][my_range].values
    predicted_values = df[col2][my_range].values

    # print(true_values[0:50])

    true_values[true_values == 0] = 0.0001

    error = np.abs(true_values - predicted_values)

    # percent_error = error/true_values

    mean_absolute_error = np.mean(error)

    print(f"Mean Absolute Error: {mean_absolute_error}")

    # print(difference[0:50])

    
    plt.plot(time, true_values, label=col1)
    plt.plot(time, predicted_values, label=col2, linestyle='dashed')
    plt.title("Joint Angle vs Time")
    plt.xlabel("Time")
    plt.ylabel("Joint Angle")
    plt.legend()
    plt.show()

    


# plt.plot(time_array, GHR_array)
# plt.show()

# print(time_array[4])
```
